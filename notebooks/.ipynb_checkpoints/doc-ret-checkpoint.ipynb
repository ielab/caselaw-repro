```python
def load_missing_file(path: str, summed=False):
    out = {}
    with open(path, 'r', encoding='utf-8') as f:
        for line in f: 
            parts = line.strip().split()
            out[parts[0]] = set(parts[1:])
            
    return out
```

```python
def load_qrel(path: str, rel: bool=False):
    qrels = {}
    with open(path) as f:
        for line in f:
            parts = line.strip().split()
            if int(parts[0]) not in qrels:
                qrels[int(parts[0])] = set()
            if rel:
                if parts[3] == '1':
                    qrels[int(parts[0])].add(parts[2])
            else:
                qrels[int(parts[0])].add(parts[2])
            
    return qrels
```

```python
def load_numret_file(path: str):
    ret = {}
    with open(path) as f:
        for line in f:
            parts = line.strip().split()
            ret[int(parts[0])] = int(parts[1])
            
    return ret 
```

```python
def load_queries(path: str):  
    queries = {}
    with open(path) as f:
        data = json.load(f)
        for topic in data['topics']:
            queries[topic['id']] = {'topic': topic['topic'], 'type': topic['type']}
    return queries
```
