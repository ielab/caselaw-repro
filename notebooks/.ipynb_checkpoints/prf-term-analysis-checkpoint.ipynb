```python
def read_file(path: str): 
    rows = []
    with open(path) as f:
        for line in f:
            parts = line.split()
            l = len(parts)
            rows.append({
                'qry': parts[0],
                'term': ' '.join(parts[1:l-12]),
                'doc': int(parts[l-12]),
                'tf': int(parts[l-11]), 
                'df': float(parts[l-10]), 
                'idf': float(parts[l-9]), 
                'recip_rank': float(parts[l-8]),
                'recall_20': float(parts[l-7]),  
                'recall_100': float(parts[l-6]), 
                'P_100': float(parts[l-5]), 
                'ndcg': float(parts[l-4]), 
                'rbp@0.10': float(parts[l-3]), 
                'rbp@0.50': float(parts[l-2]), 
                'rbp@0.80': float(parts[l-1]),
            })
    return rows

def read_nn_file(path: str): 
    rows = []
    with open(path) as f:
        for line in f:
            parts = line.split()
            l = len(parts)
            rows.append({
                'qry': parts[0],
                'term': parts[1],
                'exp_term': ' '.join(parts[2:l-13]),
                'idf': float(parts[l-13]), 
                'ictf': float(parts[l-12]), 
                'nn': int(parts[l-11]),
                'dist': float(parts[l-10]), 
                'qry_dist': float(parts[l-9]),
                'recip_rank': float(parts[l-8]),
                'recall_20': float(parts[l-7]),  
                'recall_100': float(parts[l-6]), 
                'P_100': float(parts[l-5]), 
                'ndcg': float(parts[l-4]), 
                'rbp@0.10': float(parts[l-3]), 
                'rbp@0.50': float(parts[l-2]), 
                'rbp@0.80': float(parts[l-1]),
            })
    return rows
```

```python
adg = all_decreased.groupby(['qry', 'term']).count()
adg_terms = adg[adg == 5].index.tolist()
```

```python
def get_cnt(gp):
    ind = gp.groupby(['qry']).count().index
    vals = gp.groupby(['qry']).count()['exp_term'].values
    return {t: c for t, c in zip(ind, vals)}

adg_cnts = get_cnt(adg)
```

```python
queries = load_queries(config.AUS_TOPIC_PATH) + load_queries(config.AUS_LEG_TOPIC_PATH)
```

```python
pct_dec = {}
for topic, tp in queries.items():
    tokens = tp['topic'].lower().replace(',', '').replace('?', '').replace('\'', '').replace('/', '').replace('’', '').split()
    tokens = [x for x in tokens if x not in stop.stop]
    div = float(adg_cnts.get(str(topic), 0.0))
    diff =  div / float(len(tokens))
    pct_dec[topic] = diff
    
pct_dec = sorted(pct_dec.items(), key=lambda x: x[1], reverse=True)
```

```python
for x in pct_dec:
    if x[1] < 1.0:
        break
    print(queries[x[0]])
```
